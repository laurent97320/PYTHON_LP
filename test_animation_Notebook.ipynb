```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML

fig1=plt.figure()
axe1=fig1.add_subplot(1,1,1)

X=np.arange(-5,5.1,0.1)
Y1=X*X
axe1.plot(X,Y1)
axe1.set_xticks([n for n in range(-5,6)])
axe1.set_xticks([n+0.5 for n in range(-5,6)], minor=True)

axe1.set_ylim(-1,30)
axe1.set_yticks([n for n in range(0,30)], minor=True)
axe1.set_yticklabels([n for n in range(0,30) if n%5!=0], minor=True, size=5, color="grey", style="italic")
marque,=axe1.plot(-5,5*5,"k+")
i=-5
droite,=axe1.plot(X,2*i*X-i*i)

def func(i):
    marque.set_xdata(i)
    marque.set_ydata(i*i)
    droite.set_ydata(2*i*X-i*i)
        

animation1=animation.FuncAnimation(fig1,func,frames=np.arange(-5,6,1),interval=1000, repeat=False)
plt.close()

HTML(animation1.to_jshtml())

#animation1.save("movie.mp4",fps=25)

#fig1.show()
```
